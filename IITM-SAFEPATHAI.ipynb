# SafePath AI — Project Initialization & Synthetic Dataset Generation

## Objective
The objective of this phase is to establish the foundational data infrastructure for SafePath AI, an AI-assisted urban road obstruction and emergency accessibility system designed for the IIT Madras Road Safety Hackathon 2026.

## Problem Context
Urban areas in India frequently face road obstruction issues caused by improperly parked vehicles blocking residential gates, emergency exits, pedestrian pathways, commercial entrances, and hospital access routes. These incidents can reduce accessibility, delay emergency response, and negatively impact road safety.

To simulate realistic urban obstruction scenarios, a structured synthetic dataset was generated representing:
- Vehicle obstruction incidents
- Urban locations
- Emergency conditions
- Severity classifications
- QR-based vehicle identification records

## Dataset Components

### 1. Incident Dataset
The incident dataset contains simulated obstruction reports including:
- Vehicle type
- Location category
- Obstruction type
- Emergency status
- Severity level
- Reported timestamp

### 2. Vehicle QR Database
The QR vehicle database simulates:
- QR-linked vehicle identities
- Masked contact information
- Registered cities
- Anonymous communication support

## Severity Classification Logic
A rule-based severity classification mechanism was implemented to identify obstruction criticality based on:
- Emergency access blockage
- Hospital and emergency locations
- Emergency flag activation
- Commercial and residential obstruction scenarios

## Outcome
The generated datasets establish the foundational infrastructure required for:
- Obstruction analytics
- AI-assisted chatbot interaction
- Emergency prioritization
- QR-based anonymous communication workflows
- Urban accessibility monitoring

```python
# ============================================================
# SAFEPATH AI - ROAD SAFETY HACKATHON 2026
# PROJECT INITIALIZATION & SYNTHETIC DATASET
# ============================================================


# ============================================================
# IMPORTING REQUIRED LIBRARIES
# ============================================================

import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)


# ============================================================
# PROJECT CONFIGURATION
# ============================================================

PROJECT_NAME = "SafePath AI"
PROJECT_THEME = "AI-Assisted Urban Road Obstruction & Emergency Accessibility System"

print("PROJECT INITIALIZED SUCCESSFULLY")
print(f"Project Name: {PROJECT_NAME}")
print(f"Theme: {PROJECT_THEME}")


# ============================================================
# DEFINING SYNTHETIC DATA PARAMETERS
# ============================================================

vehicle_types = [
    "Two Wheeler",
    "Car",
    "Auto Rickshaw",
    "Mini Truck",
    "Van",
    "Taxi"
]

locations = [
    "Apartment Gate",
    "Hospital Entrance",
    "Residential Street",
    "Commercial Complex",
    "School Zone",
    "Narrow Lane",
    "Pedestrian Sidewalk",
    "Shop Entrance",
    "Traffic Junction",
    "Emergency Exit"
]

cities = [
    "Chennai",
    "Bangalore",
    "Hyderabad",
    "Mumbai",
    "Delhi",
    "Kochi",
    "Pune"
]

severity_levels = [
    "Low",
    "Medium",
    "Critical"
]

obstruction_types = [
    "Gate Blockage",
    "Illegal Parking",
    "Emergency Access Blocked",
    "Pedestrian Path Obstructed",
    "Commercial Entrance Blocked",
    "Roadside Congestion"
]

emergency_flags = [
    "Yes",
    "No"
]


# ============================================================
# GENERATING SYNTHETIC INCIDENT DATASET
# ============================================================

incident_records = []

for i in range(1, 501):

    incident_id = f"INC{i:04d}"

    vehicle = random.choice(vehicle_types)

    location = random.choice(locations)

    city = random.choice(cities)

    obstruction = random.choice(obstruction_types)

    emergency_status = random.choice(emergency_flags)

    timestamp = datetime.now() - timedelta(
        minutes=random.randint(1, 10000)
    )

    # ========================================================
    # RULE-BASED SEVERITY CLASSIFICATION
    # ========================================================

    if obstruction == "Emergency Access Blocked":
        severity = "Critical"

    elif location in ["Hospital Entrance", "Emergency Exit"]:
        severity = "Critical"

    elif emergency_status == "Yes":
        severity = "Critical"

    elif obstruction in ["Gate Blockage", "Commercial Entrance Blocked"]:
        severity = "Medium"

    else:
        severity = "Low"

    incident_records.append([
        incident_id,
        vehicle,
        location,
        city,
        obstruction,
        emergency_status,
        severity,
        timestamp
    ])


# ============================================================
# CREATING INCIDENT DATAFRAME
# ============================================================

incident_df = pd.DataFrame(
    incident_records,
    columns=[
        "Incident_ID",
        "Vehicle_Type",
        "Location",
        "City",
        "Obstruction_Type",
        "Emergency_Status",
        "Severity_Level",
        "Reported_Time"
    ]
)


# ============================================================
# GENERATING SYNTHETIC QR VEHICLE DATABASE
# ============================================================

vehicle_records = []

for i in range(1, 301):

    qr_id = f"QR{i:04d}"

    owner_id = f"USER{i:04d}"

    vehicle_number = f"TN{random.randint(10,99)}AB{random.randint(1000,9999)}"

    masked_contact = f"98XXXX{random.randint(1000,9999)}"

    registered_city = random.choice(cities)

    vehicle_records.append([
        qr_id,
        owner_id,
        vehicle_number,
        masked_contact,
        registered_city
    ])


vehicle_df = pd.DataFrame(
    vehicle_records,
    columns=[
        "QR_ID",
        "Owner_ID",
        "Vehicle_Number",
        "Masked_Contact",
        "Registered_City"
    ]
)


# ============================================================
# DISPLAYING DATASET OVERVIEW
# ============================================================

print("\n")
print("="*70)
print("INCIDENT DATASET OVERVIEW")
print("="*70)

print(incident_df.head())

print("\n")
print("="*70)
print("VEHICLE QR DATABASE OVERVIEW")
print("="*70)

print(vehicle_df.head())


# ============================================================
# DATASET DIMENSIONS
# ============================================================

print("\n")
print("="*70)
print("DATASET DIMENSIONS")
print("="*70)

print(f"Incident Dataset Shape: {incident_df.shape}")
print(f"Vehicle Dataset Shape: {vehicle_df.shape}")


# ============================================================
# CHECKING MISSING VALUES
# ============================================================

print("\n")
print("="*70)
print("MISSING VALUE ANALYSIS")
print("="*70)

print(incident_df.isnull().sum())

print("\n")
print(vehicle_df.isnull().sum())


# ============================================================
# SEVERITY DISTRIBUTION
# ============================================================

print("\n")
print("="*70)
print("SEVERITY LEVEL DISTRIBUTION")
print("="*70)

print(
    incident_df["Severity_Level"]
    .value_counts()
)


# ============================================================
# EXPORTING DATASETS
# ============================================================

incident_df.to_csv("safepath_incident_dataset.csv", index=False)

vehicle_df.to_csv("safepath_vehicle_database.csv", index=False)

print("\n")
print("="*70)
print("DATASETS EXPORTED SUCCESSFULLY")
print("="*70)

print("Files Created:")
print("1. safepath_incident_dataset.csv")
print("2. safepath_vehicle_database.csv")
```

# Urban Road Obstruction Hotspot Analysis

## Objective
The hotspot analysis was performed to identify urban locations that experience higher frequencies of obstruction incidents within the SafePath AI simulation environment.

## Key Observations

### 1. Residential Streets
Residential streets recorded the highest number of obstruction incidents. This highlights the growing issue of improperly parked vehicles blocking residential access points, narrow lanes, and pedestrian movement in densely populated urban environments.

### 2. School Zones
School zones emerged as one of the major obstruction-prone areas. Unauthorized parking and roadside congestion near educational institutions can significantly impact pedestrian safety, especially for children and elderly pedestrians.

### 3. Hospital Entrances & Emergency Exits
Critical obstruction incidents were frequently observed near hospital entrances and emergency exits. These locations are highly sensitive because vehicle blockage in such areas can delay emergency response and reduce accessibility during medical emergencies.

### 4. Pedestrian Sidewalks
Frequent sidewalk obstruction incidents indicate reduced pedestrian accessibility, forcing individuals onto active roads and increasing accident risk.

## System Relevance
The hotspot analysis demonstrates the practical relevance of SafePath AI in:
- identifying high-risk urban obstruction zones,
- improving emergency accessibility,
- supporting intelligent obstruction monitoring,
- enabling future smart-city integration.

## Conclusion
The analysis highlights the need for an AI-assisted urban obstruction management system capable of prioritizing emergency-sensitive locations and enabling rapid communication between affected individuals and vehicle owners.

```python
# ============================================================
#  3)- AI SEVERITY CLASSIFICATION ENGINE
# ============================================================


# ============================================================
# DEFINING PRIORITY SCORING LOGIC
# ============================================================

# AI-inspired rule-based risk scoring system

def calculate_priority_score(row):

    score = 0

    # ========================================================
    # LOCATION-BASED RISK SCORING
    # ========================================================

    critical_locations = [
        "Hospital Entrance",
        "Emergency Exit",
        "School Zone"
    ]

    medium_locations = [
        "Apartment Gate",
        "Commercial Complex",
        "Traffic Junction"
    ]

    if row["Location"] in critical_locations:
        score += 50

    elif row["Location"] in medium_locations:
        score += 30

    else:
        score += 10


    # ========================================================
    # OBSTRUCTION TYPE RISK SCORING
    # ========================================================

    critical_obstructions = [
        "Emergency Access Blocked",
        "Pedestrian Path Obstructed"
    ]

    medium_obstructions = [
        "Gate Blockage",
        "Commercial Entrance Blocked"
    ]

    if row["Obstruction_Type"] in critical_obstructions:
        score += 40

    elif row["Obstruction_Type"] in medium_obstructions:
        score += 25

    else:
        score += 10


    # ========================================================
    # EMERGENCY STATUS RISK SCORING
    # ========================================================

    if row["Emergency_Status"] == "Yes":
        score += 50

    else:
        score += 5


    return score


# ============================================================
# APPLYING PRIORITY SCORING SYSTEM
# ============================================================

incident_df["Priority_Score"] = incident_df.apply(
    calculate_priority_score,
    axis=1
)


# ============================================================
# AI PRIORITY CLASSIFICATION ENGINE
# ============================================================

def classify_priority(score):

    if score >= 120:
        return "Emergency Critical"

    elif score >= 80:
        return "High Priority"

    elif score >= 50:
        return "Medium Priority"

    else:
        return "Low Priority"


incident_df["AI_Priority_Level"] = incident_df[
    "Priority_Score"
].apply(classify_priority)


# ============================================================
# RESPONSE RECOMMENDATION ENGINE
# ============================================================

def generate_response_action(priority):

    if priority == "Emergency Critical":

        return (
            "Immediate owner notification + "
            "Emergency escalation recommended"
        )

    elif priority == "High Priority":

        return (
            "Urgent owner notification required"
        )

    elif priority == "Medium Priority":

        return (
            "Standard obstruction resolution workflow"
        )

    else:

        return (
            "Low urgency monitoring"
        )


incident_df["Recommended_Action"] = incident_df[
    "AI_Priority_Level"
].apply(generate_response_action)


# ============================================================
# DISPLAYING ENHANCED INCIDENT DATA
# ============================================================

print("\n")
print("="*70)
print("AI PRIORITIZATION ENGINE OUTPUT")
print("="*70)

print(
    incident_df[
        [
            "Incident_ID",
            "Location",
            "Obstruction_Type",
            "Emergency_Status",
            "Priority_Score",
            "AI_Priority_Level",
            "Recommended_Action"
        ]
    ].head(15)
)


# ============================================================
# PRIORITY LEVEL DISTRIBUTION
# ============================================================

priority_distribution = (
    incident_df["AI_Priority_Level"]
    .value_counts()
    .reset_index()
)

priority_distribution.columns = [
    "Priority_Level",
    "Incident_Count"
]

print("\n")
print("="*70)
print("AI PRIORITY LEVEL DISTRIBUTION")
print("="*70)

print(priority_distribution)


# ============================================================
# VISUALIZATION - AI PRIORITY DISTRIBUTION
# ============================================================

# AI Priority Classification Distribution

plt.figure(figsize=(12,6))

sns.barplot(
    data=priority_distribution,
    x="Priority_Level",
    y="Incident_Count"
)

plt.title(
    "AI-Based Incident Priority Classification",
    fontsize=16,
    fontweight='bold'
)

plt.xlabel("Priority Level", fontsize=12)
plt.ylabel("Incident Count", fontsize=12)

plt.tight_layout()

plt.show()


# ============================================================
# EMERGENCY CRITICAL INCIDENT ANALYSIS
# ============================================================

critical_cases = incident_df[
    incident_df["AI_Priority_Level"]
    == "Emergency Critical"
]

print("\n")
print("="*70)
print("EMERGENCY CRITICAL INCIDENT SAMPLE")
print("="*70)

print(
    critical_cases[
        [
            "Incident_ID",
            "Location",
            "Obstruction_Type",
            "Priority_Score",
            "Recommended_Action"
        ]
    ].head(10)
)


# ============================================================
# LOCATION-WISE PRIORITY ANALYSIS
# ============================================================

location_priority_analysis = (
    incident_df.groupby("Location")[
        "Priority_Score"
    ]
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)

location_priority_analysis.columns = [
    "Location",
    "Average_Priority_Score"
]

print("\n")
print("="*70)
print("LOCATION-WISE PRIORITY SCORE ANALYSIS")
print("="*70)

print(location_priority_analysis)


# ============================================================
# VISUALIZATION - LOCATION PRIORITY ANALYSIS
# ============================================================

# Location-Based Risk Priority Analysis

plt.figure(figsize=(14,7))

sns.barplot(
    data=location_priority_analysis,
    x="Location",
    y="Average_Priority_Score"
)

plt.title(
    "Location-wise Average AI Priority Scores",
    fontsize=16,
    fontweight='bold'
)

plt.xlabel("Urban Location", fontsize=12)
plt.ylabel("Average Priority Score", fontsize=12)

plt.xticks(rotation=25)

plt.tight_layout()

plt.show()


# ============================================================
# EXPORTING AI PRIORITIZATION REPORT
# ============================================================

incident_df.to_csv(
    "enhanced_ai_priority_incident_dataset.csv",
    index=False
)

priority_distribution.to_csv(
    "priority_distribution_analysis.csv",
    index=False
)

location_priority_analysis.to_csv(
    "location_priority_analysis.csv",
    index=False
)

print("\n")
print("="*70)
print("AI PRIORITIZATION REPORTS EXPORTED SUCCESSFULLY")
print("="*70)

print("Generated Files:")
print("1. enhanced_ai_priority_incident_dataset.csv")
print("2. priority_distribution_analysis.csv")
print("3. location_priority_analysis.csv")
```

```python
# ============================================================
# 4) - QR-BASED ANONYMOUS COMMUNICATION ENGINE
# ============================================================


# ============================================================
# GENERATING QR-INCIDENT MAPPING
# ============================================================

# Randomly assigning QR IDs to incidents

incident_df["Assigned_QR_ID"] = np.random.choice(
    vehicle_df["QR_ID"],
    size=len(incident_df)
)


# ============================================================
# MERGING INCIDENT DATA WITH VEHICLE DATABASE
# ============================================================

merged_df = incident_df.merge(
    vehicle_df,
    left_on="Assigned_QR_ID",
    right_on="QR_ID",
    how="left"
)


# ============================================================
# DISPLAYING MERGED SYSTEM OVERVIEW
# ============================================================

print("\n")
print("="*70)
print("QR-BASED VEHICLE INCIDENT MAPPING")
print("="*70)

print(
    merged_df[
        [
            "Incident_ID",
            "Assigned_QR_ID",
            "Vehicle_Number",
            "Location",
            "AI_Priority_Level",
            "Masked_Contact"
        ]
    ].head(15)
)


# ============================================================
# ANONYMOUS MESSAGE GENERATION ENGINE
# ============================================================

def generate_anonymous_message(row):

    if row["AI_Priority_Level"] == "Emergency Critical":

        return (
            f"[CRITICAL ALERT] Your vehicle is causing "
            f"an obstruction near {row['Location']}. "
            f"Immediate attention required."
        )

    elif row["AI_Priority_Level"] == "High Priority":

        return (
            f"[HIGH PRIORITY] Your vehicle is currently "
            f"blocking access near {row['Location']}. "
            f"Please respond urgently."
        )

    elif row["AI_Priority_Level"] == "Medium Priority":

        return (
            f"[NOTICE] Your vehicle may be causing "
            f"access inconvenience near {row['Location']}."
        )

    else:

        return (
            f"[INFO] Minor obstruction reported "
            f"near {row['Location']}."
        )


merged_df["Anonymous_Notification"] = merged_df.apply(
    generate_anonymous_message,
    axis=1
)


# ============================================================
# DISPLAYING GENERATED NOTIFICATIONS
# ============================================================

print("\n")
print("="*70)
print("ANONYMOUS NOTIFICATION ENGINE OUTPUT")
print("="*70)

print(
    merged_df[
        [
            "Incident_ID",
            "Vehicle_Number",
            "AI_Priority_Level",
            "Anonymous_Notification"
        ]
    ].head(15)
)


# ============================================================
# EMERGENCY ESCALATION DETECTION
# ============================================================

merged_df["Emergency_Escalation"] = np.where(
    merged_df["AI_Priority_Level"] == "Emergency Critical",
    "Escalation Required",
    "Standard Workflow"
)


# ============================================================
# ESCALATION ANALYSIS
# ============================================================

escalation_analysis = (
    merged_df["Emergency_Escalation"]
    .value_counts()
    .reset_index()
)

escalation_analysis.columns = [
    "Escalation_Status",
    "Incident_Count"
]

print("\n")
print("="*70)
print("EMERGENCY ESCALATION ANALYSIS")
print("="*70)

print(escalation_analysis)


# ============================================================
# VISUALIZATION - ESCALATION ANALYSIS
# ============================================================

# Emergency Escalation Requirement Analysis

plt.figure(figsize=(10,5))

sns.barplot(
    data=escalation_analysis,
    x="Escalation_Status",
    y="Incident_Count"
)

plt.title(
    "Emergency Escalation Requirement Distribution",
    fontsize=16,
    fontweight='bold'
)

plt.xlabel("Escalation Status", fontsize=12)
plt.ylabel("Incident Count", fontsize=12)

plt.tight_layout()

plt.show()


# ============================================================
# COMMUNICATION LOG GENERATION
# ============================================================

communication_logs = merged_df[
    [
        "Incident_ID",
        "Assigned_QR_ID",
        "Vehicle_Number",
        "Masked_Contact",
        "AI_Priority_Level",
        "Emergency_Escalation",
        "Anonymous_Notification"
    ]
].copy()


communication_logs["Communication_Status"] = np.where(
    communication_logs["AI_Priority_Level"] == "Emergency Critical",
    "Priority Alert Sent",
    "Notification Sent"
)


# ============================================================
# DISPLAYING COMMUNICATION LOGS
# ============================================================

print("\n")
print("="*70)
print("COMMUNICATION LOG SYSTEM")
print("="*70)

print(
    communication_logs.head(15)
)


# ============================================================
# PRIORITY-WISE COMMUNICATION ANALYSIS
# ============================================================

priority_communication_analysis = (
    communication_logs["AI_Priority_Level"]
    .value_counts()
    .reset_index()
)

priority_communication_analysis.columns = [
    "Priority_Level",
    "Communication_Count"
]

print("\n")
print("="*70)
print("PRIORITY-WISE COMMUNICATION ANALYSIS")
print("="*70)

print(priority_communication_analysis)


# ============================================================
# VISUALIZATION - PRIORITY COMMUNICATION ANALYSIS
# ============================================================

# Priority Communication Distribution

plt.figure(figsize=(12,6))

sns.barplot(
    data=priority_communication_analysis,
    x="Priority_Level",
    y="Communication_Count"
)

plt.title(
    "Priority-wise Anonymous Communication Distribution",
    fontsize=16,
    fontweight='bold'
)

plt.xlabel("Priority Level", fontsize=12)
plt.ylabel("Communication Count", fontsize=12)

plt.tight_layout()

plt.show()


# ============================================================
# SAMPLE REAL-TIME INCIDENT SIMULATION
# ============================================================

sample_case = merged_df.sample(1)

print("\n")
print("="*70)
print("LIVE INCIDENT RESPONSE SIMULATION")
print("="*70)

print(
    sample_case[
        [
            "Incident_ID",
            "Location",
            "Vehicle_Number",
            "AI_Priority_Level",
            "Anonymous_Notification",
            "Emergency_Escalation"
        ]
    ]
)


# ============================================================
# EXPORTING COMMUNICATION SYSTEM REPORTS
# ============================================================

merged_df.to_csv(
    "safepath_integrated_system_dataset.csv",
    index=False
)

communication_logs.to_csv(
    "communication_logs.csv",
    index=False
)

escalation_analysis.to_csv(
    "emergency_escalation_analysis.csv",
    index=False
)

print("\n")
print("="*70)
print("COMMUNICATION SYSTEM REPORTS EXPORTED SUCCESSFULLY")
print("="*70)

print("Generated Files:")
print("1. safepath_integrated_system_dataset.csv")
print("2. communication_logs.csv")
print("3. emergency_escalation_analysis.csv")
```

```python
# ============================================================
# 5)- AI CHATBOT INTERACTION ENGINE
# ============================================================


# ============================================================
# CHATBOT RESPONSE ENGINE
# ============================================================

def generate_chatbot_response(location, obstruction_type):

    # ========================================================
    # EMERGENCY-SENSITIVE LOCATIONS
    # ========================================================

    critical_locations = [
        "Hospital Entrance",
        "Emergency Exit",
        "School Zone"
    ]

    critical_obstructions = [
        "Emergency Access Blocked",
        "Pedestrian Path Obstructed"
    ]


    # ========================================================
    # CRITICAL CASE DETECTION
    # ========================================================

    if (
        location in critical_locations
        or obstruction_type in critical_obstructions
    ):

        return {
            "Priority": "Emergency Critical",
            "Chatbot_Response":
                (
                    "Critical obstruction detected. "
                    "Immediate notification has been "
                    "triggered to the vehicle owner. "
                    "Emergency escalation workflow activated."
                ),
            "Recommended_Action":
                (
                    "Urgent clearance required."
                )
        }


    # ========================================================
    # MEDIUM PRIORITY CASES
    # ========================================================

    elif obstruction_type in [
        "Gate Blockage",
        "Commercial Entrance Blocked"
    ]:

        return {
            "Priority": "High Priority",
            "Chatbot_Response":
                (
                    "High-priority obstruction detected. "
                    "Anonymous owner notification initiated."
                ),
            "Recommended_Action":
                (
                    "Vehicle owner response requested."
                )
        }


    # ========================================================
    # STANDARD CASES
    # ========================================================

    else:

        return {
            "Priority": "Medium Priority",
            "Chatbot_Response":
                (
                    "Obstruction report registered successfully. "
                    "Standard communication workflow initiated."
                ),
            "Recommended_Action":
                (
                    "Monitoring and notification enabled."
                )
        }


# ============================================================
# SIMULATED CHATBOT INCIDENT QUERIES
# ============================================================

sample_chatbot_queries = pd.DataFrame({

    "Reported_Location": [
        "Hospital Entrance",
        "Residential Street",
        "School Zone",
        "Commercial Complex",
        "Emergency Exit",
        "Pedestrian Sidewalk"
    ],

    "Reported_Obstruction": [
        "Emergency Access Blocked",
        "Illegal Parking",
        "Gate Blockage",
        "Commercial Entrance Blocked",
        "Emergency Access Blocked",
        "Pedestrian Path Obstructed"
    ]
})


# ============================================================
# APPLYING CHATBOT RESPONSE ENGINE
# ============================================================

chatbot_outputs = []

for index, row in sample_chatbot_queries.iterrows():

    chatbot_result = generate_chatbot_response(
        row["Reported_Location"],
        row["Reported_Obstruction"]
    )

    chatbot_outputs.append({

        "Location":
            row["Reported_Location"],

        "Obstruction_Type":
            row["Reported_Obstruction"],

        "AI_Priority":
            chatbot_result["Priority"],

        "Chatbot_Response":
            chatbot_result["Chatbot_Response"],

        "Recommended_Action":
            chatbot_result["Recommended_Action"]
    })


chatbot_df = pd.DataFrame(chatbot_outputs)


# ============================================================
# DISPLAYING CHATBOT OUTPUTS
# ============================================================

print("\n")
print("="*70)
print("AI CHATBOT INTERACTION OUTPUT")
print("="*70)

print(chatbot_df)


# ============================================================
# CHATBOT PRIORITY ANALYSIS
# ============================================================

chatbot_priority_analysis = (
    chatbot_df["AI_Priority"]
    .value_counts()
    .reset_index()
)

chatbot_priority_analysis.columns = [
    "Priority_Level",
    "Incident_Count"
]

print("\n")
print("="*70)
print("CHATBOT PRIORITY ANALYSIS")
print("="*70)

print(chatbot_priority_analysis)


# ============================================================
# VISUALIZATION - CHATBOT PRIORITY DISTRIBUTION
# ============================================================

# Chatbot-Based Incident Classification

plt.figure(figsize=(10,5))

sns.barplot(
    data=chatbot_priority_analysis,
    x="Priority_Level",
    y="Incident_Count"
)

plt.title(
    "AI Chatbot Incident Priority Classification",
    fontsize=16,
    fontweight='bold'
)

plt.xlabel("Priority Level", fontsize=12)
plt.ylabel("Incident Count", fontsize=12)

plt.tight_layout()

plt.show()


# ============================================================
# RECOMMENDED ACTION ANALYSIS
# ============================================================

recommended_action_analysis = (
    chatbot_df["Recommended_Action"]
    .value_counts()
    .reset_index()
)

recommended_action_analysis.columns = [
    "Recommended_Action",
    "Count"
]

print("\n")
print("="*70)
print("RECOMMENDED ACTION ANALYSIS")
print("="*70)

print(recommended_action_analysis)


# ============================================================
# VISUALIZATION - ACTION ANALYSIS
# ============================================================

# Chatbot Recommendation Distribution

plt.figure(figsize=(14,6))

sns.barplot(
    data=recommended_action_analysis,
    x="Recommended_Action",
    y="Count"
)

plt.title(
    "AI Chatbot Recommended Action Distribution",
    fontsize=16,
    fontweight='bold'
)

plt.xlabel("Recommended Action", fontsize=12)
plt.ylabel("Count", fontsize=12)

plt.xticks(rotation=15)

plt.tight_layout()

plt.show()


# ============================================================
# REAL-TIME CHATBOT SIMULATION
# ============================================================

print("\n")
print("="*70)
print("REAL-TIME CHATBOT INCIDENT SIMULATION")
print("="*70)


test_case = generate_chatbot_response(
    "Hospital Entrance",
    "Emergency Access Blocked"
)

print("\nUSER INPUT:")
print(
    "A vehicle is blocking emergency access "
    "near the hospital entrance."
)

print("\nCHATBOT RESPONSE:")
print(test_case["Chatbot_Response"])

print("\nRECOMMENDED ACTION:")
print(test_case["Recommended_Action"])

print("\nPRIORITY LEVEL:")
print(test_case["Priority"])


# ============================================================
# EXPORTING CHATBOT REPORTS
# ============================================================

chatbot_df.to_csv(
    "chatbot_interaction_dataset.csv",
    index=False
)

chatbot_priority_analysis.to_csv(
    "chatbot_priority_analysis.csv",
    index=False
)

recommended_action_analysis.to_csv(
    "chatbot_recommended_actions.csv",
    index=False
)

print("\n")
print("="*70)
print("CHATBOT SYSTEM REPORTS EXPORTED SUCCESSFULLY")
print("="*70)

print("Generated Files:")
print("1. chatbot_interaction_dataset.csv")
print("2. chatbot_priority_analysis.csv")
print("3. chatbot_recommended_actions.csv")
```

```python
# ============================================================
# 7) - MACHINE LEARNING PREDICTION ENGINE
# ============================================================


# ============================================================
# IMPORTING MACHINE LEARNING LIBRARIES
# ============================================================

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import LabelEncoder

from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score
)

import seaborn as sns


# ============================================================
# CREATING ML DATASET COPY
# ============================================================

ml_df = incident_df.copy()


# ============================================================
# FEATURE ENGINEERING
# ============================================================

# Creating numerical emergency feature

ml_df["Emergency_Flag"] = (
    ml_df["Emergency_Status"]
    .map({
        "Yes": 1,
        "No": 0
    })
)


# Creating severity encoding score

severity_score_mapping = {

    "Low": 1,
    "Medium": 2,
    "Critical": 3
}

ml_df["Severity_Score"] = (
    ml_df["Severity_Level"]
    .map(severity_score_mapping)
)


# ============================================================
# LABEL ENCODING CATEGORICAL FEATURES
# ============================================================

location_encoder = LabelEncoder()

vehicle_encoder = LabelEncoder()

obstruction_encoder = LabelEncoder()

target_encoder = LabelEncoder()


ml_df["Location_Encoded"] = (
    location_encoder.fit_transform(
        ml_df["Location"]
    )
)

ml_df["Vehicle_Encoded"] = (
    vehicle_encoder.fit_transform(
        ml_df["Vehicle_Type"]
    )
)

ml_df["Obstruction_Encoded"] = (
    obstruction_encoder.fit_transform(
        ml_df["Obstruction_Type"]
    )
)

ml_df["Priority_Target"] = (
    target_encoder.fit_transform(
        ml_df["AI_Priority_Level"]
    )
)


# ============================================================
# FEATURE SELECTION
# ============================================================

X = ml_df[[

    "Location_Encoded",

    "Vehicle_Encoded",

    "Obstruction_Encoded",

    "Emergency_Flag",

    "Severity_Score"
]]

y = ml_df["Priority_Target"]


# ============================================================
# TRAIN TEST SPLIT
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(

    X,

    y,

    test_size=0.25,

    random_state=42,

    stratify=y
)


print("\n")
print("="*70)
print("TRAIN TEST SPLIT INFORMATION")
print("="*70)

print(f"Training Dataset Size: {X_train.shape[0]}")
print(f"Testing Dataset Size: {X_test.shape[0]}")


# ============================================================
# RANDOM FOREST MODEL INITIALIZATION
# ============================================================

rf_model = RandomForestClassifier(

    n_estimators=150,

    max_depth=8,

    random_state=42
)


# ============================================================
# MODEL TRAINING
# ============================================================

rf_model.fit(

    X_train,

    y_train
)


print("\n")
print("="*70)
print("MODEL TRAINING COMPLETED SUCCESSFULLY")
print("="*70)


# ============================================================
# MODEL PREDICTIONS
# ============================================================

y_pred = rf_model.predict(X_test)


# ============================================================
# MODEL ACCURACY EVALUATION
# ============================================================

model_accuracy = accuracy_score(

    y_test,

    y_pred
)


print("\n")
print("="*70)
print("MODEL ACCURACY SCORE")
print("="*70)

print(
    f"Random Forest Accuracy: "
    f"{round(model_accuracy*100,2)}%"
)


# ============================================================
# CLASSIFICATION REPORT
# ============================================================

target_class_names = (
    target_encoder.classes_
)

classification_results = classification_report(

    y_test,

    y_pred,

    target_names=target_class_names
)


print("\n")
print("="*70)
print("CLASSIFICATION REPORT")
print("="*70)

print(classification_results)


# ============================================================
# CONFUSION MATRIX GENERATION
# ============================================================

conf_matrix = confusion_matrix(

    y_test,

    y_pred
)


print("\n")
print("="*70)
print("CONFUSION MATRIX")
print("="*70)

print(conf_matrix)


# ============================================================
# CONFUSION MATRIX HEATMAP VISUALIZATION
# ============================================================

plt.figure(figsize=(10,7))

sns.heatmap(

    conf_matrix,

    annot=True,

    fmt='d',

    cmap='Blues',

    xticklabels=target_class_names,

    yticklabels=target_class_names
)

plt.title(
    "AI Incident Classification Confusion Matrix",
    fontsize=16,
    fontweight='bold'
)

plt.xlabel("Predicted Labels", fontsize=12)

plt.ylabel("Actual Labels", fontsize=12)

plt.tight_layout()

plt.show()


# ============================================================
# FEATURE IMPORTANCE ANALYSIS
# ============================================================

feature_importance = pd.DataFrame({

    "Feature": X.columns,

    "Importance": rf_model.feature_importances_
})


feature_importance = (
    feature_importance
    .sort_values(
        by="Importance",
        ascending=False
    )
)


print("\n")
print("="*70)
print("FEATURE IMPORTANCE ANALYSIS")
print("="*70)

print(feature_importance)


# ============================================================
# FEATURE IMPORTANCE VISUALIZATION
# ============================================================

plt.figure(figsize=(12,6))

sns.barplot(

    data=feature_importance,

    x="Importance",

    y="Feature",

    palette="viridis"
)

plt.title(
    "Machine Learning Feature Importance Analysis",
    fontsize=16,
    fontweight='bold'
)

plt.xlabel("Importance Score", fontsize=12)

plt.ylabel("Feature", fontsize=12)

plt.tight_layout()

plt.show()


# ============================================================
# MODEL PREDICTION SAMPLE OUTPUT
# ============================================================

prediction_sample = X_test.copy()

prediction_sample["Actual_Label"] = (
    target_encoder.inverse_transform(y_test)
)

prediction_sample["Predicted_Label"] = (
    target_encoder.inverse_transform(y_pred)
)


print("\n")
print("="*70)
print("MODEL PREDICTION SAMPLE OUTPUT")
print("="*70)

print(
    prediction_sample.head(15)
)


# ============================================================
# LINE CHART - INCIDENT TREND SIMULATION
# ============================================================

trend_data = (
    incident_df["AI_Priority_Level"]
    .value_counts()
    .reset_index()
)

trend_data.columns = [
    "Priority_Level",
    "Incident_Count"
]


plt.figure(figsize=(10,6))

plt.plot(

    trend_data["Priority_Level"],

    trend_data["Incident_Count"],

    marker='o',

    linewidth=3
)

plt.title(
    "Urban Incident Priority Trend Analysis",
    fontsize=16,
    fontweight='bold'
)

plt.xlabel("Priority Level", fontsize=12)

plt.ylabel("Incident Count", fontsize=12)

plt.grid(True)

plt.tight_layout()

plt.show()


# ============================================================
# PIE CHART - ESCALATION DISTRIBUTION
# ============================================================

escalation_distribution = (
    merged_df["Emergency_Escalation"]
    .value_counts()
)


plt.figure(figsize=(8,8))

plt.pie(

    escalation_distribution,

    labels=escalation_distribution.index,

    autopct='%1.1f%%',

    startangle=90
)

plt.title(
    "Emergency Escalation Distribution",
    fontsize=16,
    fontweight='bold'
)

plt.tight_layout()

plt.show()


# ============================================================
# EXPORTING ML REPORTS
# ============================================================

feature_importance.to_csv(
    "ml_feature_importance.csv",
    index=False
)

prediction_sample.to_csv(
    "ml_prediction_sample_output.csv",
    index=False
)

print("\n")
print("="*70)
print("MACHINE LEARNING REPORTS EXPORTED SUCCESSFULLY")
print("="*70)

print("Generated Files:")
print("1. ml_feature_importance.csv")
print("2. ml_prediction_sample_output.csv")
```

```python
# IMPORT REQUIRED LIBRARIES

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import IsolationForest

plt.style.use('ggplot')
```

```python
# LOAD DATASETS

incident_df = pd.read_csv("safepath_integrated_system_dataset.csv")

print("="*70)
print("SAFEPATH SMART CITY ANALYTICS ENGINE INITIALIZED")
print("="*70)

print("\nDataset Shape:")
print(incident_df.shape)

print("\nDataset Columns:")
print(incident_df.columns.tolist())
```

```python
# ADVANCED FEATURE ENGINEERING

location_encoder = LabelEncoder()
priority_encoder = LabelEncoder()
obstruction_encoder = LabelEncoder()

incident_df['Location_Encoded'] = location_encoder.fit_transform(
    incident_df['Location']
)

incident_df['Priority_Encoded'] = priority_encoder.fit_transform(
    incident_df['AI_Priority_Level']
)

incident_df['Obstruction_Encoded'] = obstruction_encoder.fit_transform(
    incident_df['Obstruction_Type']
)

# CREATE EMERGENCY FLAG

incident_df['Emergency_Flag'] = np.where(
    incident_df['AI_Priority_Level'].isin([
        'Emergency Critical',
        'High Priority'
    ]),
    1,
    0
)

# CREATE SEVERITY SCORE

severity_mapping = {
    'Low': 1,
    'Medium': 2,
    'Critical': 3
}

incident_df['Severity_Score'] = incident_df['Severity_Level'].map(
    severity_mapping
)

# CREATE ACCESSIBILITY RISK INDEX

incident_df['Accessibility_Risk_Index'] = (
    incident_df['Priority_Score'] * 0.5 +
    incident_df['Emergency_Flag'] * 30 +
    incident_df['Severity_Score'] * 10
)

print("="*70)
print("ADVANCED FEATURE ENGINEERING OUTPUT")
print("="*70)

print(
    incident_df[[
        'Incident_ID',
        'Location',
        'AI_Priority_Level',
        'Emergency_Flag',
        'Severity_Score',
        'Accessibility_Risk_Index'
    ]].head(15)
)
```

```python
# SMART CITY CLUSTER ANALYSIS USING KMEANS

cluster_features = incident_df[[
    'Location_Encoded',
    'Priority_Score',
    'Severity_Score',
    'Accessibility_Risk_Index'
]]

kmeans = KMeans(n_clusters=4, random_state=42)

incident_df['Urban_Cluster'] = kmeans.fit_predict(cluster_features)

print("="*70)
print("SMART CITY CLUSTERING OUTPUT")
print("="*70)

print(incident_df[['Incident_ID',
                   'Location',
                   'AI_Priority_Level',
                   'Urban_Cluster']].head(20))
```

```python
# CLUSTER DISTRIBUTION ANALYSIS

cluster_distribution = incident_df['Urban_Cluster'].value_counts().reset_index()

cluster_distribution.columns = ['Cluster_ID', 'Incident_Count']

print("="*70)
print("URBAN CLUSTER DISTRIBUTION")
print("="*70)

print(cluster_distribution)
```

```python
# VISUALIZATION — URBAN INCIDENT CLUSTER DISTRIBUTION

plt.figure(figsize=(10,6))

sns.barplot(
    x='Cluster_ID',
    y='Incident_Count',
    data=cluster_distribution,
    palette='viridis'
)

plt.title("Urban Incident Intelligence Cluster Distribution",
          fontsize=18,
          fontweight='bold')

plt.xlabel("Urban Intelligence Cluster")
plt.ylabel("Incident Count")

plt.tight_layout()
plt.show()
```

```python
# ACCESSIBILITY RISK TREND ANALYSIS

risk_trend = incident_df.groupby('Location')['Accessibility_Risk_Index'].mean().reset_index()

risk_trend = risk_trend.sort_values(
    by='Accessibility_Risk_Index',
    ascending=False
)

print("="*70)
print("ACCESSIBILITY RISK TREND ANALYSIS")
print("="*70)

print(risk_trend)
```

```python
# VISUALIZATION — ACCESSIBILITY RISK TREND

plt.figure(figsize=(14,7))

sns.lineplot(
    x='Location',
    y='Accessibility_Risk_Index',
    data=risk_trend,
    marker='o',
    linewidth=3,
    color='darkblue'
)

plt.title("Urban Accessibility Risk Trend Analysis",
          fontsize=18,
          fontweight='bold')

plt.xlabel("Urban Location")
plt.ylabel("Average Accessibility Risk Index")

plt.xticks(rotation=25)

plt.tight_layout()
plt.show()
```

```python
# ANOMALY DETECTION ENGINE USING ISOLATION FOREST

anomaly_features = incident_df[[
    'Priority_Score',
    'Severity_Score',
    'Accessibility_Risk_Index'
]]

iso_model = IsolationForest(
    contamination=0.05,
    random_state=42
)

incident_df['Anomaly_Flag'] = iso_model.fit_predict(anomaly_features)

incident_df['Anomaly_Status'] = incident_df['Anomaly_Flag'].map({
    1: 'Normal Incident',
    -1: 'Critical Urban Anomaly'
})

print("="*70)
print("AI ANOMALY DETECTION ENGINE OUTPUT")
print("="*70)

print(incident_df[['Incident_ID',
                   'Location',
                   'Accessibility_Risk_Index',
                   'Anomaly_Status']].head(20))
```

```python
# ANOMALY DISTRIBUTION ANALYSIS

anomaly_distribution = incident_df['Anomaly_Status'].value_counts().reset_index()

anomaly_distribution.columns = ['Anomaly_Status', 'Incident_Count']

print("="*70)
print("ANOMALY DISTRIBUTION ANALYSIS")
print("="*70)

print(anomaly_distribution)
```

```python
# VISUALIZATION — ANOMALY DETECTION DISTRIBUTION

plt.figure(figsize=(8,8))

plt.pie(
    anomaly_distribution['Incident_Count'],
    labels=anomaly_distribution['Anomaly_Status'],
    autopct='%1.1f%%',
    startangle=140,
    colors=['skyblue', 'orange']
)

plt.title("Critical Urban Anomaly Detection Distribution",
          fontsize=18,
          fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# SMART LOCATION HOTSPOT HEATMAP

heatmap_data = incident_df.pivot_table(
    values='Accessibility_Risk_Index',
    index='Location',
    columns='AI_Priority_Level',
    aggfunc='mean'
)

print("="*70)
print("SMART LOCATION HEATMAP DATA")
print("="*70)

print(heatmap_data)
```

```python
# VISUALIZATION — SMART CITY ACCESSIBILITY HEATMAP

plt.figure(figsize=(12,8))

sns.heatmap(
    heatmap_data,
    annot=True,
    cmap='YlOrRd',
    linewidths=0.5
)

plt.title("Smart City Accessibility Risk Heatmap",
          fontsize=18,
          fontweight='bold')

plt.xlabel("Priority Level")
plt.ylabel("Urban Location")

plt.tight_layout()
plt.show()
```

```python
# AI-DRIVEN PREDICTIVE GOVERNANCE INSIGHTS

governance_insights = []

top_risk_location = risk_trend.iloc[0]['Location']

governance_insights.append(
    f"Highest accessibility risk hotspot identified: {top_risk_location}"
)

critical_anomalies = anomaly_distribution[
    anomaly_distribution['Anomaly_Status'] == 'Critical Urban Anomaly'
]['Incident_Count'].values[0]

governance_insights.append(
    f"Total critical urban anomalies detected: {critical_anomalies}"
)

largest_cluster = cluster_distribution.iloc[0]['Cluster_ID']

governance_insights.append(
    f"Urban Cluster {largest_cluster} emerged as the dominant incident intelligence zone."
)

governance_insights.append(
    "AI clustering successfully grouped urban obstruction patterns."
)

governance_insights.append(
    "Predictive anomaly detection identified potential emergency-sensitive incidents."
)

print("="*70)
print("PREDICTIVE GOVERNANCE INSIGHTS")
print("="*70)

for idx, insight in enumerate(governance_insights, start=1):
    print(f"{idx}. {insight}")
```

```python
# EXPORT ADVANCED ANALYTICS REPORTS

cluster_distribution.to_csv(
    "urban_cluster_distribution.csv",
    index=False
)

risk_trend.to_csv(
    "accessibility_risk_trend_analysis.csv",
    index=False
)

anomaly_distribution.to_csv(
    "urban_anomaly_distribution.csv",
    index=False
)

incident_df.to_csv(
    "advanced_safepath_city_intelligence_dataset.csv",
    index=False
)

print("="*70)
print("ADVANCED SMART CITY ANALYTICS REPORTS EXPORTED SUCCESSFULLY")
print("="*70)

print("\nGenerated Files:")
print("1. urban_cluster_distribution.csv")
print("2. accessibility_risk_trend_analysis.csv")
print("3. urban_anomaly_distribution.csv")
print("4. advanced_safepath_city_intelligence_dataset.csv")
```
